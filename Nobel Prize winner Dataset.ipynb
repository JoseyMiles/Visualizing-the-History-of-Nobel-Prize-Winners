```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("nobel.csv")
data.head()
```

```python
#Top gender
top_gender= data.value_counts("sex").index[0]
top_gender
```

```python
num_male_female = data["sex"].value_counts()
num_male_female
```

```python
#Top country
top_country= data.value_counts("birth_country").index[0]
top_country
```

```python
#Column for winners with birth_country "United States of America"
data["USA"]= data["birth_country"]== "United States of America"
data.head()
```

```python
#Column for decades
data["decades"] = (data["year"] //10)*10
data.head()
```

```python
#Decade with winners from USA
decade_usa= data.groupby("decades", as_index=False)["USA"].mean()
decade_usa
```

```python
#Maximum decade in USA
max_decade_usa = decade_usa[decade_usa["USA"]==decade_usa["USA"].max()]["decades"].values[0]
max_decade_usa
```

```python
#Filtering for female winners
data["female_winner"]= data["sex"]== "Female"
data.head()
```

```python
#Decade, Category with average proportion of females
avg_prop_females= data.groupby(["decades", "category"], as_index=False)["female_winner"].mean()
avg_prop_females
```

```python
#Dictionary
max_female_dict = {
    avg_prop_females[avg_prop_females['female_winner'] == avg_prop_females['female_winner'].max()]['decades'].values[0]: avg_prop_females[avg_prop_females['female_winner'] == avg_prop_females['female_winner'].max()]['category'].values[0]
}
max_female_dict
```

```python
#Data frame showing the rows where women won the nobel prize
women = data[data["female_winner"]]
women.head()
```

```python
#Filtering women by the earliest year
earliest_year = women[women["year"]== women["year"].min()]
earliest_year
```

```python
#First woman to win the nobel price
first_woman_name = earliest_year["full_name"].values[0]
print(first_woman_name)
```

```python
#First woman category
first_woman_category = earliest_year["category"].values[0]
print(first_woman_category)
```

```python
#Repeated names
counts= data.groupby(["full_name"], as_index=True).size().reset_index(name='count')
counts
```

```python
#Greater than one time
greater_than_one = counts[counts["count"]>=2].set_index("full_name").index
greater_than_one
```

```python
#converting to list
repeat_list = list(greater_than_one)
repeat_list
```
